# Advanced removal
Similarities were found with tf-idf representation and cosine similarity. Since thus approach is time and memory consuming, it was performed externally and resluts are imported here.  
Warning: dask index is not very reliable, so it is better to drop dupplicates by index with pandas

```python
max_sim = 0.92

sim = pd.read_csv('sim-h-80.csv')
sim = sim.loc[sim.similarity > max_sim]
grouped = sim.groupby(['I1'])['I2'].apply(list).reset_index(name='duplicates')
ids_to_be_removed = set()
for l in list(grouped.duplicates):
    ids_to_be_removed.update(l)
```

```python
dataset_pd = pd.read_csv('0-datasets/v-1.0.0-beta-h/dataset-0.csv')
dataset_pd = dataset_pd.drop(list(ids_to_be_removed))
dataset_pd.describe()
```
